```python
import pandas as pd

# ---------- 在此处手动填入文件路径及对应的国家 ----------
# 格式：每个元组为 (路径, 国家名)
file_country_pairs = [
    ("/home/ninini/Datasets_Nine_country/Brazil/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "Brazil"),
    ("/home/ninini/Datasets_Nine_country/Indonesia/llm/gpt-4o/sensitive_words_cleaned.csv", "Indonesia"),
    ("/home/ninini/Datasets_Nine_country/Mexico/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "Mexico"),
    ("/home/ninini/Datasets_Nine_country/Saudi-Arabia/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "Saudi-Arabia"),
    ("/home/ninini/Datasets_Nine_country/Singapore/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "Singapore"),
    ("/home/ninini/Datasets_Nine_country/South-Africa/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "South-Africa"),
    ("/home/ninini/Datasets_Nine_country/Thailand/llm/deepseek-v4-flash/sensitive_words_cleaned.csv", "Thailand"),
    ("/home/ninini/Datasets_Nine_country/Turkiye/llm/deepseek-v4-flash/sensitive_words_cleaned.csv","Turkiye"),
    # ... 继续添加
]

# 读取并合并
df_list = []
for file_path, country in file_country_pairs:
    df = pd.read_csv(file_path)                 # 内部需有 word, category, meaning 列
    df['country'] = country
    df_list.append(df)

merged = pd.concat(df_list, ignore_index=True)

# 按国家和类别聚类排序
merged = merged.sort_values(['country', 'category']).reset_index(drop=True)

# 保存
merged.to_csv("merged_sensitive_words.csv", index=False)

print("合并完成，共 {} 条记录".format(len(merged)))
print(merged.head())
```
